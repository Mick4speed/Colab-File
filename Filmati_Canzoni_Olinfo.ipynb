Codice:

Ci sono 2 modi per risolvere questo problema, uno è svillupato con un ciclo e quindi ha complessita *O(N)* e un altro lineare con complessità *O(1)*, per ora ci interessa sapere che è preferibile usare il secondo metodo, anche se in questo caso non cambia di tanto perché gli input sono picolli.

1) Esempio: Usiamo un ciclo *while* e ad ogni tornata togliamo o un filmato o un CD, rispetto allo spazio rimasto, fino a quando non finiamo lo spazio sul disco

````python
%%writefile download.cpp
// NOTA: si raccomanda di usare questo template anche se non lo si capisce completamente.

#include <fstream>
#include <iostream>
#include <string>
#include <vector>

using namespace std;

int main() {
    // decommenta le due righe seguenti se vuoi leggere/scrivere da file
    ifstream cin("input.txt");
    ofstream cout("output.txt");

    int T;
    cin >> T;
    for (int test = 1; test <= T; ++test) {
        int N, F, C;
        cin >> N >> F >> C;

        int nf = 0, nc = 0;


        while(N>0){
          if(F<=N){
            N-=F;
            nf++;
          }
          else if(C<=N){
            N-=C;
            nc++;
          }
          else{
            break;
          }
        }


        cout << "Case #" << test << ": ";
        cout << nf << " " << nc << endl;
    }

    return 0;
}
````

````python
%%bash
g++ download.cpp -o download.o
./download.o
cat output.txt
````

2) Esempio: prima troviamo il numero di filmato calcolando *N/F*, poi togliamo da *N* i byte dei filmati scaricati, e poi  troviamo il numero di CD calcolando *N/C*

````python
%%writefile download_test.cpp
#include <string>
#include <iostream>
#include <fstream>

using namespace std;

int main() {
    // se preferisci leggere e scrivere da file
    // ti basta decommentare le seguenti due righe:

    freopen("input.txt", "r", stdin);
    freopen("output.txt", "w", stdout);

    int T, t;
    cin >> T;

    for (t = 1; t <= T; t++) {
        int N, F, C;
        cin>>N>>F>>C;

        int nf=N/F;
        N-=nf*F;
        int nc = N/C;

        cout << "Case #" << t << ": " << nf << " " << nc << endl;
    }
}
````

````python
%%bash
g++ download_test.cpp -o download.o
./download.o
cat output.txt
````

Questo approccio è più efficiente perché indipendentemente dai valori della variabili saranno eseguiti lo stesso numero di istruzioni

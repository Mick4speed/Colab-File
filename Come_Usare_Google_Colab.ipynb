# ***Ordinare ArrayList***

Con il metodo 'sort' noi possiamo ordinare un arrayList, e passando il parametro 'String.CASE_INSENSITIVE_ORDER' lo ordiniamo senza fare distinzione tra maiuscole e minuscole.

```python
%%writefile order_arraylist.jsh
import java.util.ArrayList;

ArrayList<String> nomi= new ArrayList<>();
nomi.add("Michael");
nomi.add("angelo");
nomi.add("Jerald");
nomi.add("clemente");

System.out.println("nomi: "+nomi);
nomi.sort(String.CASE_INSENSITIVE_ORDER);
System.out.println("nomi ordinati: "+nomi);
/exit
```

```python
!jshell order_arraylist.jsh
```

Invece se vogliamo ordinarli rispettando maiuscolo e minuscolo useremo la funzione 'Collections.sort(*nome_arrayList*);'

```python
%%writefile order_arraylist2.jsh
import java.util.ArrayList;

ArrayList<String> nomi= new ArrayList<>();
nomi.add("Michael");
nomi.add("angelo");
nomi.add("Jerald");
nomi.add("clemente");

System.out.println("nomi: "+nomi);
Collections.sort(nomi);
System.out.println("nomi ordinati: "+nomi);
/exit
```

Da notare che ordina prima le maiuscole e poi le minuscole
